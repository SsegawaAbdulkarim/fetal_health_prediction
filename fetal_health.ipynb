```python
X = data.drop("fetal_health", axis='columns')
y = data.fetal_health
```

```python
col = X.columns
col
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
X = scaler.fit_transform(X)
```

```python
import pandas as pd
X = pd.DataFrame(data=X, columns = col)
X.head()
```

```python
from sklearn.feature_selection import SelectKBest, chi2
selector = SelectKBest(score_func=chi2, k=6).fit(X, y)
```

```python
selected_cols = X.columns[selector.get_support()]
selected_cols
```

```python
selected_cols
```

```python
best_cols = X[selected_cols]
best_cols.head()
```

```python
from sklearn.model_selection import train_test_split
x_trian, x_test, y_train, y_test = train_test_split(best_cols, y, test_size=0.3)
x_trian.shape, x_test.shape
```
